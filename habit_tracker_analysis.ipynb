```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("habit_tracker.csv")
df.head()
```

```python
# Convertir la columna DATE a formato datetime
df["DATE"] = pd.to_datetime(df["DATE"], errors="coerce")


cols = ["DATE", "READ", "WORKOUT", "STUDY", "MEDITATE", "SLEEP 7h", "SCORE"]
df = df[cols]


habit_cols = ["READ", "WORKOUT", "STUDY", "MEDITATE", "SLEEP 7h"]
df[habit_cols] = df[habit_cols].replace({"TRUE": 1, "FALSE": 0})


df[habit_cols] = df[habit_cols].fillna(0)
df[habit_cols] = df[habit_cols].apply(pd.to_numeric, errors="coerce").fillna(0).astype(int)
```

```python
habit_totals = df[habit_cols].sum().sort_values(ascending=False)
habit_totals
```

```python
habit_totals.plot(kind="bar")
plt.title("Habit Completion in March")
plt.ylabel("Completions")
plt.xlabel("Habit")

plt.savefig("habit_completion.png") 
plt.show()
```

```python
plt.figure()
plt.plot(df["DATE"], df["SCORE"])
plt.title("Daily Productivity Score")
plt.xlabel("Date")
plt.ylabel("Score")

plt.savefig("productivity_trend.png")
plt.show()
```

# Habit Tracker Data Analysis

## Objective
Analyze daily habits to identify which habit is the most consistent and how productivity evolves over time.

## Dataset
This dataset contains daily records of habits such as reading, workout, study, meditation, sleep, and a productivity score.

## Habit Completion Analysis

The bar chart shows the total number of completions for each habit during the month.

Sleep (7h) is the most consistent habit, indicating strong discipline in maintaining rest.

On the other hand, meditation has the lowest completion rate, suggesting an area for improvement.

Workout and study show moderate consistency, while reading is relatively stable but not the highest.

Overall, the analysis highlights which habits are well established and which ones require more attention.

## Score Trend Analysis

The line chart represents the daily productivity score over time.

It shows how productivity fluctuates throughout the month, highlighting both high-performance and low-performance days.

This visualization helps identify patterns, such as periods of improvement or inconsistency, and provides insight into overall productivity trends.

Understanding these trends can help in adjusting habits and improving future performance.

## Conclusion

This analysis shows that sleep is the most consistent habit, suggesting a strong foundation in rest and recovery.

However, habits like meditation and study show lower consistency, indicating areas that could be improved to enhance overall productivity.

Additionally, the productivity score fluctuates significantly, which suggests that maintaining consistent habits may help stabilize performance over time.

Overall, focusing on improving weaker habits while maintaining strong ones could lead to better productivity and balance.
